-  작가 중 선정, 지숙 은 /로 나눠져있지않고 ,로 나누어져 있어서 특수문자 제거 후 분리
-  장르 카테고리(스토리, 옵니버스 등) 삭제 

```python
webtoon['author'] = webtoon['author'].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
webtoon['writer'] = webtoon['author'].apply(lambda x: x.split(' ')[0])
webtoon['illustrator'] = webtoon['author'].apply(lambda x:x.split(' ')[-1])

webtoon['genre'] = webtoon['genre'].apply(lambda x: x.split(',')[-1])
webtoon.drop(columns='author', inplace=True)
```

```python
with open('stopwords_korean.txt','r', encoding='utf-8') as f :
    stopwords = f.readline()
stopwords = stopwords.split()
print(stopwords[:10])
```
