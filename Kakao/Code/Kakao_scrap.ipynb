```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import urllib
import urllib.request
from selenium.common.exceptions import NoSuchElementException
import time
import pyautogui
import os
```

```python
#데이터 불러오기

import json

with open('weekdays.json', 'r', encoding='utf8') as f:
    contents = f.read()
    json_data = json.loads(contents)

# print(json.dumps(json_data, indent="\t", ensure_ascii=False))
# print(json_data['data']['sections'][0]['cardGroups'][0]['cards'][0]['content']['title'])

# 구조 정제

title = json_data['data']['sections'][0]['cardGroups'][0]['cards'][0]['content']['title']
keyword = json_data['data']['sections'][0]['cardGroups'][0]['cards'][0]['content']['seoKeywords']
synopsis = json_data['data']['sections'][0]['cardGroups'][0]['cards'][0]['content']['synopsis']
authors = json_data['data']['sections'][0]['cardGroups'][0]['cards'][0]['content']['authors']
featuredCharacterImageB = json_data['data']['sections'][0]['cardGroups'][0]['cards'][0]['content']['featuredCharacterImageB']

# print(title)
# print(keyword)
# print(synopsis)
# print(authors)
# print(featuredCharacterImageB+'.png')

x = 0
y = 0
z = 0

days = []
titles = []
keywords = []
synopsises = []
authorsl = []
main_image = []

while True:
  try:
    xs = json_data['data']['sections'][x]
    while True:
      try:
        ys = xs['cardGroups'][y]
        while True:
          try:
            day = xs['title']
            title = ys['cards'][z]['content']['title']
            keyword = ys['cards'][z]['content']['seoKeywords']
            synopsis = ys['cards'][z]['content']['synopsis']
            authors = ys['cards'][z]['content']['authors']
            images = ys['cards'][z]['content']['featuredCharacterImageB']

            days.append(day)
            titles.append(title)
            keywords.append(keyword)
            synopsises.append(synopsis)
            authorsl.append(authors)
            main_image.append(images)

            z += 1
          except:
            z = 0
            y += 1
            break
      except:
        z = 0
        y = 0
        x += 1
        break
  except:
    break
    
#제목 정제
import re

re_titles = []

for i in range(len(titles)):
    re_titles.append(re.sub('[^0-9a-zA-Zㄱ-힗]','',titles[i]))
    
v_titles = ['사귄건아닌데','지옥사원','말하지말까','프레너미','나만아는사랑','주말도미시식회','취향저격그녀']
for i in range(len(v_titles)):
    re_titles.remove(v_titles[i])
re_titles
```

```python
def Video_Webtoon_Crawler():
    URL = 'https://webtoon.kakao.com/original-webtoon?tab=mon'

    driver = webdriver.Chrome(executable_path='C:/chromedriver')
    driver.implicitly_wait(time_to_wait=1)
    driver.get(URL)

    v_titles = ['사귄건아닌데','지옥사원','말하지말까','프레너미','나만아는사랑','주말도미시식회','취향저격그녀']

    #Vedio 웹툰
    for i in range(1,8):

        driver.find_element_by_xpath('//*[@id="root"]/main/div/div[1]/div[1]/div[3]/div[2]/div[2]/ul/li['+str(i+1)+']/p').click()
        time.sleep(1)

        first = driver.find_element_by_xpath('//*[@id="root"]/main/div/div[1]/div[2]/div/div[3]/div/div/div/div/div/div['+str(i)+']/div[3]/div[1]/div/div/div/div/div/a/video')
        first.click()
        time.sleep(1)

        pyautogui.mouseDown()

        for iter_1 in range(1,11):
            pyautogui.scroll(-1000)
            time.sleep(1)

        for iter_2 in range(1,11):
            pyautogui.scroll(1000)
            time.sleep(1)


        img_url = driver.find_elements_by_xpath('//*[@id="root"]/main/div/div[2]/div/div[1]/div[3]/div/div/div[1]/div/div[2]/div/div[1]/div/div/div/div/ul/li/a/div[1]/div/picture/img')
    #     len(img_url)

        img_list = []

        for img in img_url:
            img_list.append(img.get_attribute('src'))

    #     print(img_list)
        time.sleep(1)

        if not os.path.exists('./imgs/{}'.format(v_titles[i-1])):
            os.mkdir('./imgs/{}'.format(re.sub('[^0-9a-zA-Zㄱ-힗]','',v_titles[i-1])))

        for k in range(len(img_url)):
            urllib.request.urlretrieve(img_list[k], './imgs/{}/{}{}.jpg'.format(v_titles[i-1], v_titles[i-1], str(k+1)+ '화'))

        driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div/div[1]/div[1]/div[1]/div[1]/img').click()
        time.sleep(1)

    driver.close()
```

```python
monday_titles = re_titles[:28] #월요일
tuesday_titles = re_titles[28:53] #화요일
wednesday_titles = re_titles[53:79] #수요일
thursday_titles = re_titles[79:107] #목요일
friday_titles = re_titles[107:133] #금요일
saturday_titles = re_titles[133:158] #토요일
sunday_titles = re_titles[158:] #일요일


#image 웹툰
def Image_Webtoon_Crawler(titles, day):
    URL = 'https://webtoon.kakao.com/original-webtoon?tab=mon'

    driver = webdriver.Chrome(executable_path='C:/chromedriver')
    driver.implicitly_wait(time_to_wait=1)

    driver.get(URL)
    time.sleep(1)

    #로그인 페이지로 이동
    driver.find_element_by_xpath('//*[@id="root"]/main/div/div/div[1]/div[2]/div[1]/img').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div[2]/div/div[1]/div/div/div/div/div/a').click()
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/div[3]/div/div/div/div[1]/div/div/button').click()
    time.sleep(1)

    #로그인
    driver.find_element_by_id("id_email_2").send_keys('아이디 입력')
    driver.find_element_by_id("id_password_3").send_keys('비밀번호 입력')
    driver.find_element_by_xpath('//*[@id="login-form"]/fieldset/div[8]/button[1]').click()
    time.sleep(1)

    #웹툰 원작 페이지로 이동
    driver.find_element_by_xpath('//*[@id="root"]/main/div/div/div[1]/div[2]/div[2]/img').click()
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/div[7]/div/div/div/a/p').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div[1]/div[3]/div[1]/div/div[8]/div/p').click()
    time.sleep(1)

    #요일 클릭
    driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div[1]/div[3]/div[2]/div[2]/ul/li['+str(day+1)+']/p').click()
    time.sleep(1)


    #xpath 변동 초기화 위해 페이지 이동
    first = driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div[2]/div/div[3]/div/div/div/div/div/div['+ str(day) +']/div[3]/div[2]/div/div/div/div/a')
    first.click()

    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div/div[1]/div[1]/div[1]/div[1]/img').click()
    time.sleep(1)

    num = 0

    try:

        for j in range(2,50):

            if j == 4:
                continue
            if j == 6:
                continue

            first = driver.find_element_by_xpath('//*[@id="root"]/main/div/div[1]/div[2]/div/div[3]/div/div/div/div/div/div['+str(day)+']/div[3]/div['+str(j)+']/div/div/div/div/a')
            first.click()
            time.sleep(1)

            pyautogui.mouseDown()

            for i in range(1,11):
                pyautogui.scroll(-1000)
                time.sleep(1)

            for i in range(1,11):
                pyautogui.scroll(1000)
                time.sleep(1)


            img_url = driver.find_elements_by_xpath('//*[@id="root"]/main/div/div[2]/div/div[1]/div[3]/div/div/div[1]/div/div[2]/div/div[1]/div/div/div/div/ul/li/a/div[1]/div/picture/img')
        #     len(img_url)

            img_list = []

            for img in img_url:
                img_list.append(img.get_attribute('src'))

        #         print(img_list)
            time.sleep(1)

            if not os.path.exists('./imgs/{}'.format(titles[num])):
                os.mkdir('./imgs/{}'.format(re.sub('[^0-9a-zA-Zㄱ-힗]','',titles[num])))

            for k in range(len(img_url)):
                urllib.request.urlretrieve(img_list[k], './imgs/{}/{}{}.jpg'.format(titles[num], titles[num], str(k+1)+ '화'))

            num+=1

            driver.find_element_by_xpath('//*[@id="root"]/main/div/div[2]/div/div[1]/div[1]/div[1]/div[1]/img').click()
            time.sleep(1)

    except:
        #로그아웃
        driver.find_element_by_xpath('//*[@id="root"]/main/div/div[1]/div[1]/div[2]/div[2]/div/a[2]/img').click()
        time.sleep(1)
        
        driver.find_element_by_xpath('//*[@id="root"]/main/div/div[3]/div[3]/div[2]/div[1]/a[1]/div/p').click()
        time.sleep(1)
        
        driver.find_element_by_xpath('//*[@id="root"]/main/div/div[3]/div[2]/div/a[2]/div/p').click()
        time.sleep(1)
        
        print('종료')

    driver.close()
```
